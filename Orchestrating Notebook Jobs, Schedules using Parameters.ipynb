```python
#fetching the librabry
from pyspark.sql.functions import *
```

```python
%python
#view data at location /databricks-datasets/retail-org/company_employees/company_employees.csv
dbutils.fs.head("/databricks-datasets/retail-org/company_employees/company_employees.csv")
```

```python
%python
#create a parameter for the dept
dbutils.widgets.text("dept", "","Department Name") 
```

```python
%python
#get a parameter value in a variable _dept
_dept=dbutils.widgets.get("dept")


print(_dept)
```

```python
%python
#read all the data through spark
df=spark.read.csv("/databricks-datasets/retail-org/company_employees/company_employees.csv",header=True)
display(df)
```

```python
#filter the records
df_filtered=df.filter((df.department==_dept.upper()) & (df.active_record==1))
```

```python
display(df_filtered)
```

```python
#Write the data to delta table with department name
_count=df_filtered.count()
if _count>0:
    df_filtered.write.format("delta").mode("overwrite").saveAsTable(f"dev.bronze.dept_{_dept}".lower())
    print(f"data is written sucessully,to {_dept.lower()},along with count of {_count}")
    display(spark.table(f"dev.bronze.dept_{_dept}".lower()))
else:
    print(f"no data to write,count is {_count}")
```

```python
#exit status along with the no of data written to table
dbutils.notebook.exit(f"data written successfully  with {_count}")
```
